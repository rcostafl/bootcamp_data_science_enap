### Motivação

Sabe-se que o Brasil possui um sistema tributário heterogêneo relativamente à base de incidência dos tributos. Alguns tributos incidem sobre o consumo (bens e serviços), outros, sobre patrimônio e renda e etc.

Os tributos sobre o consumo são chamados de tributos indiretos, não pessoais, ou seja, não incidem sobre o indivídui. Pelo fato de não levarem em conta a capacidade contributiva do contribuinte, sendo sendo que todos os contribuites pagam exatamente mesmo valor. Por este fato são considerados regressivos, na medida em que tendem o mesmo valor pago em tributo representa, proporcionalmente, uma um ônus maior, proporcionalmente, para aqueles que possuem um rendimento menor, ou seja, tributam mais, proporcionalmente, quem tem uma menor capacidade contributiva.

Contratiamente, a tributação sobre a renda é direta, incidindo sobre a renda de cada um dos indivíduos e tende a permitir a progressividade, com a majoração de alíquotas proporcionalmente ao aumento da renda. 

Entretanto, no Brasil, as suas alíquotas progressivas do IR estão concentradas na faixa de remuneração de zero até BRL 5.000,00 (aprox.), ou seja, quem recebe BRL 5.000,00/mês está sujeito à mesma alíqutoa de imposto de renda que o estão os contribuintes que recebem BRL 100 mil (ou mais) /  mês.

Além disso, por meio de publicações do índice de GINI, sabe-se que há uma grande desigualdade na distruição de renda no Brasil.

A motivação deste trabalho é tentar entender se há alguma relação entre a carga tributária, considerando-se inclusive a sua base de incidência (diretas e indiretas), e a desigualdade (concentração de renda).

```python
# YOUR CODE HERE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # Biblioteca para plotar gráficos
import seaborn as sns # Biblioteca para plotar gráficos baseada no matplotlib
sns.set() 
```

```python
##############################################################
#                      Configurações                         #
##############################################################
# Caminho para os dados gini da OCDE
gini_ocde_file_path = "../../home_work_coding_bootcamp_ds/data/gini/gini_disposable_income_post_tax_transfers.csv"

# Caminho para os dados gini da OCDE
gini_wb_file_path = "../../home_work_coding_bootcamp_ds/data/gini/gini_wb.csv" # <- Banco mundial

# informando o ano selecionado
selected_year = 2017
```

```python
##############################################################
#                             GINI                           #
##############################################################

#### Carregando dados do índice de GINI (fonte: Banco Mundial)

# Definindo os tipos das colunas de interesse.
column_types_wb={
    "Country Name":"string",
    "Country Code":"string",
    "Indicator Name": "string",
    str(selected_year): "float64" # Os valores estão nesta coluna, cujo nome é o ano.
}

# Carregando os dados do csv para a variável gini.
gini = pd.read_csv(gini_wb_file_path, skiprows=4, header=0, dtype=column_types_wb)

# Renomeando colunas do dataframe gini para facilitar o seu acesso posterior.
gini = gini.rename(
    columns={
        "Country Name": "Country", 
        "Country Code":"country_code", 
        "Indicator Name": "Measure"
    }
)

# Selecionando as colunas de interesse, para deixar o dataframe mais limpo.
gini = gini[[
    "Country",
    "country_code",
    "Measure",
    str(selected_year)
]]

#    Criando colunas index_name e Year, para facilitar eventuais consultas posteriores aos dados do dataframe.
gini["index_name"] = "GINI"                              # Adicionando coluna index_name.
gini["index_name"] = gini["index_name"].astype("string") # Definindo o tipo da coluna como string.

gini["Year"] = selected_year # Adicionando coluna Year que conterá o ano alvo.

#     Renomeando coluna que contém os valores GINI (2017) para gini_value. Como será realizado um merge
# entre os dataframes do índice d GINI e o do carga tributária (tax burden - tb), ao renomear esta coluna,
# evitar-se-á confusão entre as colunas valor de ambos os dataframes
gini.rename(columns={str(selected_year):"gini_value"}, inplace=True)
gini = gini[gini["gini_value"].notna()]

# Organizando as colunas para que tenham a mesma sequência do dataframe gini
gini = gini[["country_code", "Country","index_name", "Measure", "Year", "gini_value"]]

gini
```

```python
#################################################################################################################
#                                    CARGA TRIBUTÁRIA: ARRECADAÇÃO/PIB                                          #
#################################################################################################################

########################## Função: load_tax_income(tax_type) ##############################
#                                                                                         #
# - Descrição: carregada dados da relação arrecadação tributária/PIB, tendo como fonte    #
#   dados da OCDE.                                                                        #
#                                                                                         #
# - Parâmetro:                                                                            #
#  - Tipo: String                                                                         #
#  - Valores possíveis:                                                                   #
#    - TOTALTAX : retorna a carga tributária - CT relativa à tributação total             #
#    - 1000     : retorna a CT relativa à tributação sobre a renda                        #
#    - 2000     : retorna a tributáção da seguridade social                               #
#    - 3000     : retorna a CT relativa à tributação incidente sobre folha de pagamentos  #
#    - 4000     : retorna a CT relativa à tributação incidente sobre patrimônio           #
#    - 5000     : retorna a CT relativa à tributação incidente sobre bens e serviços      #
#    - 6000     : retorna a CT relativa à tributação incidente sobre outras bases         #
#
# - Retorno: dataframe com a carga tributária (arrecadação/PIB)                           #
###########################################################################################
def load_tax_income(tax_type):    
    # ct = carga tributária: arrecadação tributária como percentual do PIB
    
    ### Carga dos dados baixados da OCDE
    # Definição dos tipos dos dados
    data_type={
        "COU":"string", 
        "Country":"string", 
        "Level of government": "string",
        "TAX": "string",
        "VAR": "string"
    }

    # Leitura e carga do arquivo de carga tributária
    ct_percentual_pib = pd.read_csv("./data/total_tax_percentage_gdp.csv", header=0, dtype=data_type)
    ct_percentual_pib.rename(columns={"COU":"country_code"}, inplace=True)

    # Criando filtros
    flt_ano       = ct_percentual_pib["Year"] == selected_year          # Dados do ano de 2017
    flt_nivel_gov = ct_percentual_pib["Level of government"] == "Total" # Todos os níveis de governo (federal, estadual e municipal)
    flt_tributo   = ct_percentual_pib["TAX"] == tax_type                # Todos os tributos
    flt_base      = ct_percentual_pib["VAR"] == "TAXGDP"                # Somente dados da relação arrecadação/PIB (a relação poderia ser percentual da arrecadação de um tributo em relação à arrecadação total)

    # Aplicando filtros
    ct_percentual_pib = ct_percentual_pib[
        flt_ano       & # Somente dados do ano selecionado (2017)
        flt_nivel_gov & # Dados relativos a todos os níveis de governo
        flt_tributo   & # Dados relativos a todos os tributos (sobre renda, bens e serviços, folha de salário e etc.)
        flt_base       # Dados da arregadação em relação ao PIB (carga_tributária = arrecadação/PIB)
    ]     

    # Selecionando colunas de interesse
    ct_percentual_pib = ct_percentual_pib[[
        "country_code", 
        "Country", 
        "TAX", 
        "Tax revenue", 
        "Indicator", 
        "Year", 
        "Value"
    ]]
    
    ct_percentual_pib.rename(columns={"Value":"tax_burden_value"}, inplace=True)

    # Retornando o dataframe da carga tributŕia
    return ct_percentual_pib
```

```python
#################################################################################################################
#                                         CARGA TRIBUTÁRIA x GINI                                               #
#################################################################################################################
### Será realizado um merge dos dataframes gini e da carga tributária com os seguintes objetivos:
# - Selecionar somente os países que tenham dados tanto para a CT quanto para o índice de Gini
# - Montar os gráficos 
# - Verificar a correção

# Buscando os dados da carga tributária -> arrecadação total / PIB
ct_tot_percentual_pib = load_tax_income("TOTALTAX") # Carga tributária total.

#    Fazendo um join entre os dados gini e a carga tributária, para sincronizar os países com dados
# relativos à carga tributária e ao índice de Gini.
gini_ct_tot_pib = pd.merge(gini, ct_tot_percentual_pib, on="country_code")
gini_ct_tot_pib
```

Devidamente carregados os dados relativos ao índice de Gini e à carga tributária total, a análise dos países que possuíam dados para ambos os indicadores. Abaixo, seguem gráfico para o posicionamento do Brasil em relação aos demais países analisados.

Isso poderia ser explicado pelo fato de que a tributação sobre a renda pode ser progressiva, tributando mais aqueles que tem maior capacidade contributiva, recursos estes que iriam para o governo e seriam, posteriormente, distributídos, facilitando a distribuição de renda, ou seja, os mais ricos pagam mais tributos ao governo, que distribui a renda, aumentando o poder de consumo dos mais pobres.

Pelo exposto acima, fica claro que há uma maior dificuldade em se distribuir renda quando se adota uma tributação indireta em relação a quando se adota uma tributação direta, podendo esta ser uma explicação possível para a fraca correalação entre o índice de GINI e a carga tributária relativa aos tributos incidentes sobre o consumo.

Voltando à questão da discrepância entre o índice de GINI e a sua carga tributária, tendo e em conta a informação aqui apresentada, faz-se importante uma análise cmoparativa entre o quanto o Brasil explora as duas bases de tributação em questão em relação aos demais países da amostra.

```python
# Gerando o gráfico de gini
tl_graphic = gini_tb_profit
# Mostrar os gráficos do matplotlib na mesma janela do jupyter notebook
#%matplotlib inline

# Adicionando colunas estatísticas: media e mediana
df_group_data = pd.DataFrame({
    "country_code":["MDM", "MAN"], 
    "Country_x": ["Medium", "Mean"],
    "index_name": ["GINI", "GINI"],
    "Measure": ["Gini index (World Bank estimate)", "Gini index (World Bank estimate)"],
    "Year_x": [selected_year, selected_year],
    "gini_value":[tl_graphic["gini_value"].median(), tl_graphic["gini_value"].mean()],
    "Country_y":["medium", "medium"],
    "TAX":["1000", "1000"],
    "Tax revenue":["000 Taxes on income, profits and capital gains", "000 Taxes on income, profits and capital gains"],
    "Indicator":["Tax revenue as % of GDP", "Tax revenue as % of GDP"],
    "Year_y":[2017, 2017],
    "tax_burden_value":[tl_graphic["tax_burden_value"].median(), tl_graphic["tax_burden_value"].median()]
})

tl_graphic = tl_graphic.append(df_group_data)

tl_graphic = tl_graphic.sort_values(by=["tax_burden_value"])

# Define um objeto de figura
fig = plt.figure(figsize=(25,8))

# Adiciona um subplot, pois uma figura pode ter vários plots
ax = fig.add_subplot()

# Cores diferentes para Brasil, Media (MAN - MEAN) e Mediana (MDM - Medium)
colors_per_code = {
    "BRA": "green",
    "MAN": "lightgrey",    #color for mean
    "MDM": "blue" #color for medium
}

# Desenha um gráfico de linhas com o dataset
# Plota cada linha. lw significa line width
collor_pallet = [colors_per_code[c] if (c in ["BRA", "MAN", "MDM"]) else "lightblue" for c in tl_graphic["country_code"].values]
tl_plot = sns.barplot(x="country_code",y='tax_burden_value',data=tl_graphic,lw=4,palette=collor_pallet,ax=ax)

plt.title('CARGA TRIBUTÁRIA: Trib. sobre a Renda', fontsize=22, ha='center', color='black')
plt.ylabel('CARGA TRIB.',fontsize=16,color='black')
plt.xlabel('Países',fontsize=16,color='black', labelpad=10)
```

Pela análise do gráfico acima, verifica-se que a medida em que o Brasil, em verde, explora a renda para tributação é muito próximo à média (em cinza - MAN - MEAN) e à mediana (em azul - MDM - MEDIUM) da medida em que os demais países da amostra exploram esta mesma base de incidência tributária.

Faz-se necessária, neste momento, a comparativo do quanto o Brasil explora a tributação sobre o consumo (tributação indireta), em realao aos demais países da amostra.

### Conclusões e Achados

Duranta a análise dos dados apresentados, observou-se que a distribuição de renda (GINI) apresenta uma correlação relativamente forte com a carga tributária.

Apesar disso, verifiou-se que o Brasil possui a pior distribuição de renda da amostra, apesar de ter uma carga tributária muito próxima da média e da media da amostra, ou seja, há uma discrepância entre entre os indicadores .

Isso poderia ser explicado pelo fato que a carga tributária é composta por soma das cargas tributárias de diversas bases de incidência tributária (renda, seguridade social, folha de salários, patrimônio, bens e serviços e outros).

Essas bases de incidência apresentam correlações diferentes com o indicador GINI. A tributação sobre a renda, por exemplo, apresenta uma correalão relevante com este indicador, enquanto que a tributação sobre o consumo não.

Posteriormente, verificou-se que o Brasil tributa muito bens e serviços (base que apresenta baixa correlação com o GINI) que a média dos países da amostra, posicionando-se acima do terceiro quartil.

Em contrapartida, tributa a renda próximo a média dos países da amostra.

A discrepância observada entre os indicadores GINI e carga tributária total poderia estar relacionada ao fato de o Brasil tributar mais bens e serviços do que a média dos países da amostra.

Por fim, obviamente, faz-se necessária uma análise mais aprofundada dos dados, incluindo as demais bases de incidência (atividades que estão fora do escop deste trabalho pela questão do tempo), entretanto, em um primeiro momento, considerando-se a forte correlação entre tributação sobre a renda e a índice de GINI, é bastante razoável afirmar que o deslocamento da base de incidência da tributção do consumo para a renda poderia contribuir para questão da distribuição de renda no Brasil.
